# Game Idea: Number Guessing Game
1. Code for Non-AI Version: Player vs Computer


```
import random

def number_guessing_game():
    # The computer selects a random number between 1 and 100
    number_to_guess = random.randint(1, 100)
    attempts = 0
    max_attempts = 10  # Player has 10 attempts to guess the number

    print("Welcome to the Number Guessing Game!")
    print("Guess a number between 1 and 100. You have 10 attempts.")

    # Loop for the player to make guesses
    while attempts < max_attempts:
        try:
            # Prompt the player to enter their guess
            guess = int(input("Enter your guess: "))
        except ValueError:
            # Handle non-integer input
            print("Please enter a valid integer.")
            continue

        attempts += 1  # Increment the number of attempts

        if guess < number_to_guess:
            print("Too low!")  # Provide feedback if guess is too low
        elif guess > number_to_guess:
            print("Too high!")  # Provide feedback if guess is too high
        else:
            print(f"Congratulations! You guessed the number in {attempts} attempts.")
            return  # Exit the function if the number is guessed correctly

    # Inform the player if they have used all their attempts
    print(f"Sorry, you've used all your attempts. The number was {number_to_guess}.")

# Run the game
number_guessing_game()
```

## 2. Code for AI Version with Binary Search:



```
def ai_number_guessing_game():
    # Player selects a number
    print("Think of a number between 1 and 100, and I (the AI) will try to guess it.")
    low = 1
    high = 100
    attempts = 0

    # Loop until the AI guesses the number correctly
    while low <= high:
        guess = (low + high) // 2  # AI uses binary search to guess
        attempts += 1

        print(f"AI's guess is: {guess}")
        feedback = input("Enter 'h' if too high, 'l' if too low, or 'c' if correct: ").lower()

        if feedback == 'c':
            print(f"I (AI) guessed the number in {attempts} attempts!")
            return
        elif feedback == 'h':
            high = guess - 1  # If too high, reduce the upper bound
        elif feedback == 'l':
            low = guess + 1  # If too low, increase the lower bound

    print("Something went wrong!")

# Run the AI version
ai_number_guessing_game()
```

## 3. Code for BFS(Breadth-First Search) Version:



Explanation of BFS in the Number Guessing Game

The BFS algorithm is utilized to guess a number between 1 and 100 in this guessing game by systematically exploring all possible numbers:

Initialization:

The game starts by initializing a queue with all possible numbers from 1 to 100.
The queue is a deque (double-ended queue) that stores these numbers.

Guessing Loop:

The algorithm enters a loop where it continually guesses numbers from the front of the queue.
Each guess is made by removing the first number from the queue using popleft().

Feedback Handling:

After each guess, the user provides feedback:
'h' if the guess is too high.
'l' if the guess is too low.
'c' if the guess is correct.

Depending on the feedback:
If the guess is too high ('h'), the algorithm updates the queue to only include numbers less than the current guess.
If the guess is too low ('l'), it updates the queue to only include numbers greater than the current guess.

Termination:

If the guess is correct ('c'), the algorithm prints the number of attempts and ends the game.
If the queue becomes empty without finding the correct number (which shouldn't normally happen), it indicates an error.

This BFS approach ensures that all possible numbers are explored systematically, narrowing down the range based on feedback until the correct number is guessed.

```
from collections import deque

def bfs_number_guessing_game():
    # Player selects a number
    print("Think of a number between 1 and 100, and I (the AI) will try to guess it using BFS.")
    low = 1
    high = 100
    attempts = 0

    queue = deque(range(low, high + 1))  # Initialize queue with all possible numbers

    while queue:
        guess = queue.popleft()  # Get the next guess from the queue
        attempts += 1

        print(f"AI's guess is: {guess}")
        feedback = input("Enter 'h' if too high, 'l' if too low, or 'c' if correct: ").lower()

        if feedback == 'c':
            print(f"I (AI) guessed the number in {attempts} attempts!")
            return
        elif feedback == 'h':
            # Remove guesses greater than the current guess
            queue = deque(num for num in queue if num < guess)
        elif feedback == 'l':
            # Remove guesses less than the current guess
            queue = deque(num for num in queue if num > guess)

    print("Something went wrong!")

# Run the BFS version
bfs_number_guessing_game()
```

## 4. Code for DFS(Depth-First Search) Version:


Explanation of DFS in the Number Guessing Game

The DFS algorithm is used to guess a number between 1 and 100 by exploring potential numbers in a depth-first manner:

Initialization:

The game starts by initializing a stack with all possible numbers from 1 to 100.
The stack is a list that holds these numbers.

Guessing Loop:

The algorithm enters a loop where it continually guesses numbers from the top of the stack.
Each guess is made by removing the last number from the stack using pop().

Feedback Handling:

After each guess, the user provides feedback:
'h' if the guess is too high.
'l' if the guess is too low.
'c' if the guess is correct.

Depending on the feedback:

If the guess is too high ('h'), the algorithm updates the stack to only include numbers less than the current guess.
If the guess is too low ('l'), it updates the stack to only include numbers greater than the current guess.

Termination:

If the guess is correct ('c'), the algorithm prints the number of attempts and ends the game.
If the stack becomes empty without finding the correct number (which ideally shouldn't happen), it indicates an error.

In summary, the DFS approach guesses numbers by diving deep into one possible option at a time (stack-based approach), updating the list of potential guesses based on user feedback, until the correct number is identified.

```
def dfs_number_guessing_game():
    # Player selects a number
    print("Think of a number between 1 and 100, and I (the AI) will try to guess it using DFS.")
    low = 1
    high = 100
    attempts = 0

    stack = list(range(low, high + 1))  # Initialize stack with all possible numbers

    while stack:
        guess = stack.pop()  # Get the next guess from the stack
        attempts += 1

        print(f"AI's guess is: {guess}")
        feedback = input("Enter 'h' if too high, 'l' if too low, or 'c' if correct: ").lower()

        if feedback == 'c':
            print(f"I (AI) guessed the number in {attempts} attempts!")
            return
        elif feedback == 'h':
            # Remove guesses greater than the current guess
            stack = [num for num in stack if num < guess]
        elif feedback == 'l':
            # Remove guesses less than the current guess
            stack = [num for num in stack if num > guess]

    print("Something went wrong!")

# Run the DFS version
dfs_number_guessing_game()
```

# 5. Dijkstra’s Algorithm


Explanation of Dijkstra's Algorithm in the Number Guessing Game

The game uses Dijkstra's Algorithm to guess a number between 1 and 100 by efficiently narrowing down the range based on feedback:

Initialization:

The target number is input by the user for testing purposes.
The algorithm initializes parameters including low, high, and distances for each possible guess.
A priority queue (priority_queue) is used to manage guesses based on their "distance" or number of attempts.

Setup:

distances tracks the shortest path (in terms of number of guesses) to each number. Initially, all distances are set to infinity, except for the starting guess.
The priority queue starts with the initial guess (low), having a distance of 0.

Guessing Loop:

The algorithm pops the guess with the smallest distance from the priority queue.
It checks if this guess has been visited before. If not, it proceeds to make the guess.
User feedback is used to adjust the guessing range:
'h' (too high): Update the distances for all guesses lower than the current guess and push them into the priority queue.
'l' (too low): Update the distances for all guesses higher than the current guess and push them into the priority queue.
The low and high bounds are adjusted based on feedback to refine the guessing range.

Termination:

If the guess is correct ('c'), the game announces success and ends.
If the queue is exhausted without finding the correct number, it indicates an error.

Summary:

Dijkstra's Algorithm in this game helps prioritize guesses based on their "distance" from previous guesses, optimizing the search for the target number by systematically exploring the most promising guesses first.



```
import heapq

def dijkstra_number_guessing_game():
    # Player selects a number
    print("Think of a number between 1 and 100, and I (the AI) will try to guess it using Dijkstra's Algorithm.")
    target = None
    while target is None:
        try:
            target = int(input("Enter the number you are thinking of (for testing purposes, this will not be visible to the AI): "))
            if not (1 <= target <= 100):
                raise ValueError
        except ValueError:
            print("Please enter a valid number between 1 and 100.")
            target = None

    # Initialize Dijkstra's Algorithm
    low = 1
    high = 100
    attempts = 0
    visited = set()
    priority_queue = [(0, low)]  # (distance, guess)
    distances = {i: float('inf') for i in range(low, high + 1)}
    distances[low] = 0

    while priority_queue:
        current_distance, current_guess = heapq.heappop(priority_queue)
        attempts += 1

        if current_guess in visited:
            continue
        visited.add(current_guess)

        print(f"AI's guess is: {current_guess}")
        feedback = input("Enter 'h' if too high, 'l' if too low, or 'c' if correct: ").lower()

        if feedback == 'c':
            print(f"I (AI) guessed the number {target} correctly in {attempts} attempts!")
            return
        elif feedback == 'h':
            # Update distances for all guesses lower than current_guess
            for guess in range(low, current_guess):
                if guess not in visited:
                    new_distance = current_distance + 1
                    if new_distance < distances[guess]:
                        distances[guess] = new_distance
                        heapq.heappush(priority_queue, (new_distance, guess))
            high = current_guess - 1
        elif feedback == 'l':
            # Update distances for all guesses higher than current_guess
            for guess in range(current_guess + 1, high + 1):
                if guess not in visited:
                    new_distance = current_distance + 1
                    if new_distance < distances[guess]:
                        distances[guess] = new_distance
                        heapq.heappush(priority_queue, (new_distance, guess))
            low = current_guess + 1

    print("Something went wrong!")

# Run the Dijkstra's Algorithm version
dijkstra_number_guessing_game()
```
